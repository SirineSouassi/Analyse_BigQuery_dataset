```python
# Fonction pour créer une nouvelle colonne "browser" à partir de la colonne "Device"
def creat_column_browser(data):
  list_browser = []
  for i in data.index:
    list_browser.append(data['device'][i]['web_info']['browser'])
  data['browser'] = list_browser
  return data

# Fonction pour créer une list de noms de navigateurs unique
def name_browser(list):
  list_browser_names = []
  for i in list.index:
    list_browser_names.append(i)
  return list_browser_names
```

```python
# Le nombre de vues par utilisateur unique de la page d’accueil.
# Dans cette exemple je vais travailler seulement par 31 Janvier 2021, pour être plus rapide. 

df_31_2021 = build_data()
df_31_2021 = df_31_2021.append(pd.io.gbq.read_gbq('''SELECT * From bigquery-public-data.ga4_obfuscated_sample_ecommerce.events_20201231 LIMIT 2000''',project_id = project_id, dialect = 'standard'),ignore_index = True)
```

```python
# Dans le dataset , certaines lignes ne contiennent pas de titre de page , nous allons donc les supprimer.
def remove_row_without_page_title(data):
  n = 0
  list_remove_id = [] 
  for i in data.index:
    j = 0
    test = False
    while j < len(data['event_params'][i]) and (test == False):
      if data['event_params'][i][j]['key'] == "page_title":
        n = n+1
        test = True
      j=j+1
    if test ==  False: 
      list_remove_id.append(i)
  return list_remove_id
```

```python
list_remove_id = remove_row_without_page_title(df_31_2021)

# build new data
new_df_31_2021 = df_31_2021

# Remove the rows without page_title
for i in list_remove_id:
  new_df_31_2021 = new_df_31_2021.drop(i)
```

```python
# Créer une colonne Page_title
def create_column_page_title(new_data):
  list_page_title = [] 
  for i in new_data.index:
    for j in range(0,len(new_data['event_params'][i])):
      if new_data['event_params'][i][j]['key'] == "page_title":
        list_page_title.append(new_data['event_params'][i][j]['value']['string_value'])
  new_data['page_title'] = list_page_title
  return new_data
```

```python
# Create page title column
new_df_31_2021 = create_column_page_title(new_df_31_2021)

# Extracter l'utilisateur qui ont vues le page d'accueil
x_data = new_df_31_2021.loc[(new_df_31_2021['page_title'] == 'Home') & (new_df_31_2021['event_name'] == 'page_view')]
user_id = x_data.groupby('user_pseudo_id',as_index=False)['user_pseudo_id'].first()
nb_user = x_data.groupby('user_pseudo_id', as_index=False)['user_pseudo_id'].count()
```

# **Les 10 pays qui ont le nombre maximum d'utilisateurs**




```python
# Load data
df_2021_01 = load_data_2021()
df_2020_12 = load_data_2020_12()
df_2020_11 = load_data_2020_11()
```

```python
# create data Frame 
data = []
data = pd.DataFrame(data)
data = pd.concat([df_2021_01,df_2020_12,df_2020_11], ignore_index=True)
```

```python
# Créer colonne contenant les noms de pays
def creat_column_country(data):
  list_country = []
  for i in data.index:
    list_country.append(data['geo'][i]['country'])
  data['country'] = list_country
  return data

def name_country(list):
  list_country_names = []
  for i in list.index:
    list_country_names.append(i)
  return list_country_names
```

```python
# La part d'utilisateurs selon le pays

# Ajouter une colonne qui contient les noms des navigateurs 
data = creat_column_country(data)

# Creér list qui contient les noms des pays unique
list_country_names = name_country((data['country'].value_counts()).head(10))
fig = px.bar( x = data['country'].value_counts().head(10), y = list_country_names,title = "les 10 pays qui ont le nombre maximum d'utilisateurs" , labels={
                     "x": "pays",
                     "y": "Le nombre d'utilisateurs"
                 })
fig.show()
```
